```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('BerkeleyData.csv')
```

```python
df=df1.T
df1
```

```python
Rat_FreqM=[];Rat_FreqF=[];Tot_AccptRM=[];Tot_AccptRF=[];
arr_df1=range(0,len(df1[0:]),4)
for i in arr_df1:
    print(i)
    Rat_FreqM.append(df[i][3]/(df[i][3]+df[i+1][3]))
    Rat_FreqF.append(df[i+2][3]/(df[i+2][3]+df[i+3][3]))
    Tot_AccptRM.append(df[i][3]/(df[i][3]+df[i+1][3]+df[i+2][3]+df[i+3][3]))
    Tot_AccptRF.append(df[i+2][3]/(df[i][3]+df[i+1][3]+df[i+2][3]+df[i+3][3]))
```

```python
Xticks=np.array(['A','B','C','D','E','F']);
plt.figure(figsize=(6, 6), dpi=80, facecolor='w', edgecolor='k');
plt.xlabel('Department')
plt.ylabel('Gender-based Applicant Acceptance ratio')
#plt.legend('Male','Female')
plt.plot(Xticks,Rat_FreqM,label='Male Acceptance')
plt.plot(Xticks,Rat_FreqF,label='Female Acceptance')
#plt.set_position([box.x0, box.y0, box.width * 0.8, box.height])
legend = plt.legend(shadow=True, fontsize='x-large')
legend.get_frame().set_facecolor('#00BBCC')
plt.show()
```

```python
Xticks=np.array(['A','B','C','D','E','F']);
plt.figure(figsize=(6, 6), dpi=80, facecolor='w', edgecolor='k');
plt.xlabel('Department')
plt.ylabel('Total Applicant Acceptance ratio')
#plt.legend('Male','Female')
plt.plot(Xticks,Tot_AccptRM,label='Male Acceptance')
plt.plot(Xticks,Tot_AccptRF,label='Female Acceptance')
#plt.set_position([box.x0, box.y0, box.width * 0.8, box.height])
legend = plt.legend(shadow=True, fontsize='x-large')
legend.get_frame().set_facecolor('#00BBCC')
plt.show()
```
